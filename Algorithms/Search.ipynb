```python
import math
import os
import random
import re
import sys

from collections import deque

'''
Given a chess board size n

Calculate the number of moves required for a knight that can make any sequence of [(1:n-1),(1:n-1)] moves to reach location (n-1,n-1) on the chessboard. Return -1 if not possible.

Used BFS while counting depth to count all moves
'''
def knightlOnAChessboard(n):   
    movesArr = []

    for i in range(1,n):
        movesRow = []
        for j in range(1,n):
            visited = set()
            toVisit = deque()
            toVisit.append((0,0))
            toVisit.append(None)

            val = bfs(n, i, j, visited, toVisit)
            movesRow.append(val)
        movesArr.append(movesRow)

    return movesArr

def bfs(n, xMove, yMove, visited, toVisit):

    moveCount = 0
    while(toVisit):

        coords = toVisit.popleft()
        #print(coords)
        if(coords is None):
            moveCount += 1
            toVisit.append(None)
            if(toVisit[0]==None):
                return -1
            else:
                continue

        if((coords[0],coords[1]) in visited):
            return -1

        if(coords == (n-1,n-1)):
            return moveCount

        visited.add((coords[0],coords[1]))

        possibleMoves = getPossibleMoves(n,coords,xMove,yMove,visited)
        
        for move in possibleMoves:
            if move not in toVisit:
                toVisit.append(move)

        #print(coords,":",toVisit)
        

    return moveCount


def getPossibleMoves(n,coords,xMove,yMove, visited):
    moves = []
    for xDirMove in [-xMove, xMove]:
        xnew1 = coords[0] + xDirMove
        ynew2 = coords[1] + xDirMove
        for yDirMove in [-yMove, yMove]:
            ynew1 = coords[1] + yDirMove
            xnew2 = coords[0] + yDirMove

            if(checkBoundary(n,xnew1,ynew1) and ((xnew1, ynew1) not in visited)):
                moves.append((xnew1,ynew1))
            if(checkBoundary(n,xnew2,ynew2) and ((xnew2, ynew2) not in visited) and ((xnew2,ynew2) not in moves)):
                moves.append((xnew2,ynew2))

    return moves

def checkBoundary(n,x,y):
    if((x >= 0 and x < n)and (y >= 0 and y < n)):
        return True
    return False
```

```python
class Solution(object):
    def FindCombinations(self,nums):
        combSet = self.combs(nums,[],set())
        return combSet
    
    def combs(self, nums, curr, combSet):
        
        if(len(nums) == 0):
            combSet.add(tuple(curr))
            return combSet
        
        for i in range(len(nums)):
            
            if(len(curr) != 0):
                combSet.add(tuple(curr))

            nextArr = curr + [nums[i]]
            restArr = nums[i+1:]
            self.combs(restArr,nextArr,combSet)
            
        return combSet
```

```python
class Solution(object):
    def findClosestElements(self, arr, k, x):
        """
        :type arr: List[int]
        :type k: int
        :type x: int
        :rtype: List[int]
        """
        
        #Step 1: Find the index of the element in the array closest to x
        # Ex. 1 -> index 2
        # Ex. 2 -> index 0
        
        pointerInitIndex = None
        for i in range(len(arr)):
            if(arr[i] >= x): #Found upper bound
                if(i == 0):
                    pointerInitIndex = 0
                else:
                    if(abs(arr[i]-x) < abs(arr[i-1]-x)): #if current index val closer to x
                        pointerInitIndex = i
                    else:
                        pointerInitIndex = i-1
                break
        
        if pointerInitIndex is None: pointerInitIndex = len(arr)-1 
        
        #Step 2: Create 2 pointers: one that travels left (descending), one that travels right (ascending)
        lowIndex = highIndex = pointerInitIndex
        
        countPointerMoves = 0
        
        #Step 3: Move the pointers out, until when have moved them total k times
        while(countPointerMoves < k-1):
            if(highIndex == len(arr)-1 or abs(arr[lowIndex-1]-x) <= abs(arr[highIndex + 1]-x)):
                lowIndex -= 1
            else: #high index val is closer
                highIndex += 1
                
            countPointerMoves += 1
                
        #Step 4: Splice the array at the pointer locations
        return arr[lowIndex:highIndex+1]
            
        
        
```
